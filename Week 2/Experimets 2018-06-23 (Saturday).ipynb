```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
x = np.arange(0,100)
y = np.sin(x/10)+np.random.random_integers(low=-2, high=2, size=100)/5 + 8
```

```python
plt.scatter(x,y)
```

```python
df = pd.DataFrame({"x":x,"y":y})
```

```python
df.head()
```

```python
lm.fit(df[["x"]],df.y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
plt.plot(x,y)
plt.scatter(np.arange(0,100,2),lm.predict( pd.DataFrame({"x":np.arange(0,100,2)})),c="red")
```

```python
x = np.arange(-50,70,5)
y = np.arange(-50,70,5)
x_ = []
y_ = []
z_ = []
for i in x:
    for j in y:
        x_.append(i)
        y_.append(j)
        z_.append(2*(i**2) + 3 * i * j + 4 * i + 5) # Find coef = 2,3,4, intercept = 5
```

```python
df = pd.DataFrame({"x":x_, "y": y_, "z": z_})
df.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["x"], df["y"], df["z"], c="blue", marker="o")
plt.show()
```

```python
lm = LinearRegression()
lm.fit(df.drop("z",axis=1),df.z)
print(lm.coef_)
print(lm.intercept_)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["x"], df["y"], df["z"], c="blue", marker="o")
ax.scatter(df["x"], df["y"], lm.predict(df.drop("z", axis=1)), c="red", marker=".")
plt.show()
```
