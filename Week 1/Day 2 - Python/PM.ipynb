```python
import matplotlib.pyplot as plt
import matplotlib.colors as clrs
import numpy as np
import math
```

```python
x = np.linspace(0, math.pi*2, 50)
y = np.sin(x)
```

```python
plt.plot(x,y,"g*-")
plt.show()
```

```python
plt.plot(x, y, "r")
plt.xlabel("X Axis Title Here")
plt.ylabel("Y Axis Title Here")
plt.title("String Title Here")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(x, y, "g*-")
```

```python
fig = plt.figure()
fig1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig2 = fig.add_axes([0.7, 0.5, 0.3, 0.3])
fig1.plot(x,y,"r")
fig2.plot(x,y,"g")
```

```python
plt.plot(x,y,color=clrs.rgb2hex((1,0.5,0.5)))
plt.plot(x+1,-y, color="#8B008B")
plt.show()
```

```python
plt.plot(x,y,color="purple", lw = 2, ls = "-", marker="o", markersize=6)
plt.plot(x+1,y,color="blue", lw = 2, ls = "-", marker=".", markersize=6)
plt.plot(x+2,y,color="cyan", lw = 2, ls = "-", marker="^", markersize=6)
plt.plot(x+3,y,color="green", lw = 2, ls = "-", marker="o", markersize=6)
```

```python
x = np.arange(0,100)
y = x
plt.plot(x, y)
```

```python
fig = plt.figure()
fig1 = fig.add_axes([0.0, 0.0, 1, 1])
fig2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
fig2.set_xticks(np.linspace(0,1,6))
fig2.set_yticks(np.linspace(0,1,6))
```

```python
fig = plt.figure()
fig1 = fig.add_axes([0.0, 0.0, 1, 1])
fig2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
fig2.set_xticks(np.linspace(0,100,6))
fig2.set_yticks(np.linspace(0,200,5))

fig1.plot(x,y)
fig2.plot(x,y)
fig1.set_xlabel("X")
fig1.set_ylabel("Y")
fig2.set_xlabel("X")
fig2.set_ylabel("Y")
fig.show()
```

```python
fig = plt.figure()
fig1 = fig.add_axes([0.0, 0.0, 1, 1])
fig2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
fig2.set_xticks(np.linspace(0,100,6))
fig2.set_yticks(np.linspace(0,200,5))

fig2.plot([20, 22], [40, 44])
fig2.set_ylim([30,50])
z = np.arange(0,100)
e = z ** 2
fig1.plot(z,e)
fig2.set_xlabel("X")
fig2.set_title("zoom")
fig2.set_ylabel("Z")
fig1.set_xlabel("X")
fig1.set_ylabel("Y")
fig2.set_xticks(np.linspace(20,22,5))
fig2.set_yticks(np.linspace(30,50,5))
fig.show()
```
